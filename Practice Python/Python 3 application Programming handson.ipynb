# Handson-Python-Calendar

```python
import calendar
from collections import Counter

def usingcalendar(datetuple):
    if calendar.isleap(datetuple[0]):
        lst=list(datetuple)
        lst[1]=2
        datetuple=tuple(lst)
    print(calendar.month(datetuple[0],datetuple[1]))
    obj=calendar.Calendar()
    l=[]
    for day in obj.itermonthdates(datetuple[0],datetuple[1]):
        l.append(day)
    print(l[-7:])
    count=Counter(d.strftime('%A') for d in obj.itermonthdates(datetuple[0],datetuple[1]) if d.month==datetuple[1])
    common=count.most_common(1)
    print(common[0][0])
    

if __name__ == '__main__':
    qw1 = []

    for _ in range(3):
        qw1_item = int(input().strip())
        qw1.append(qw1_item)
        
    tup=tuple(qw1)

    usingcalendar(tup)
```
